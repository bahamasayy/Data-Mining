```python
#Load library yang akan digunakan
#sklearn.svm import SVR digunakan untuk mengimport class library untuk svr
import pandas as pd
from openpyxl import load_workbook
from sklearn import preprocessing as pre
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error
```

```python
#Mengatur nama file csv yang digunakan
df = pd.read_csv('data.csv')
df
```

```python
#inisialisasi atau pendefinisian varieble dengan nama kolom
df.columns = ['date', 'price', 'bdroom', 'bathroom', 'sqft_living', 'sqt_lot', 'floors', 'waterfront', 'view', 'condition', 'sqt_above', 'sqt_basement', 'yr_build', 'yr_renovate', 'street', 'city', 'statezip', 'country']
df
```

```python
#checking tipe data
print(df.dtypes)
```

```python
#Ubah index data frame agar lebih memudahkan dalam melakukan forecasting
d_train = df [['date', 'price', 'bdroom', 'bathroom', 'sqft_living', 'sqt_lot', 'floors', 'waterfront', 'view', 'condition', 'sqt_above', 'sqt_basement', 'yr_build', 'yr_renovate']].reset_index(drop = True)
d_train
```

```python
#Menentukan kolom yang digunakan sebagai class target forcasting
#Kolom yang digunakan adalah price
y = targets = labels = d_train['price'].apply(float).values
y
```

```python
#Menentukan kolom yang digunakan sebagai fitur-fitur
columns = ['bdroom', 'bathroom', 'sqft_living', 'sqt_lot', 'floors', 'waterfront', 'view', 'condition', 'sqt_above', 'sqt_basement', 'yr_build', 'yr_renovate']
features = d_train[list(columns)].values
features
```

```python
#Menyiapkan data feature untuk proses training
scaler = pre.StandardScaler().fit(features)
X = scaler.transform(features)
X
```

```python
SVR_model = SVR(kernel = 'linear' , C = 100, gamma = 'auto').fit(X, y)
SVR_model
```

```python
#Menggunakan 100 data sample yang diambil secara random
d_test = d_train.sample(n = 100)
d_test
```

```python
#Menentukan kolom sebagai kelas target atau label pada data uji
columns = ['bdroom', 'bathroom', 'sqft_living', 'sqt_lot', 'floors', 'waterfront', 'view', 'condition', 'sqt_above', 'sqt_basement', 'yr_build', 'yr_renovate']
features_test = d_test[list(columns)].values
```

```python
#Menentukan kolom yang digunakan untuk feature pada data uji
y_true = d_test['price'].values
y_true
```

```python
#Menyiapkan data untuk proses testing
scaler_test = pre.StandardScaler().fit(features_test)
testdata = scaler_test.transform(features_test)
```

```python
#Prediksi house price 
predict = SVR_model.predict(testdata)
predict
```

```python
#Nilai Mean Absolute Precentage Error
import numpy as np

def mean_absolute_percentage_error(y_true, predict):
    y_true, predict = np.array(y_true), np.array(predict)
    return np.mean(np.abs((y_true - predict)) * 100)

mean_absolute_percentage_error(y_true, predict)
```
